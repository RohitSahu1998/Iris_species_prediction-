# Data Preparation

1) Identifying the inputs (X) and output (y) <br>
2) Split into train and test <br>
3) Preprocess the training data <br>
4) Preprocess the test data <br>

```python
df.head()
```

```python
y=df['target']
X=df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
```

```python
y.head()
```

```python
X.head()
```

```python
# splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Rescaling the Data into single scale
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_transformed=scaler.fit_transform(X_train)
print(X_train_transformed.shape)
```

```python
X_test_transformed=scaler.transform(X_test)
print(X_test_transformed.shape)
```

# Logistic Regression

```python
# Training
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train_transformed,y_train)
```

```python
# Prediction
y_test_pred=classifier.predict(X_test_transformed)
```

```python
# Evaluation
from sklearn import metrics
metrics.accuracy_score(y_test_pred,y_test)
```

# Decision Tree Classifier

```python
# Training
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train_transformed, y_train)

```

```python
# Prediction
y_test_pred=classifier.predict(X_test_transformed)
```

```python
# Evaluating
metrics.accuracy_score(y_test_pred,y_test)
```

# Random Forest Classifier

```python
# Train
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_transformed,y_train)
```

```python
# prediction
y_test_predict=classifier.predict(X_test_transformed)
```

```python
# Evaluating
metrics.accuracy_score(y_test_predict,y_test)
```

# K-Nearest Neighbors

```python
# Train
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train_transformed, y_train)
```

```python
# Predict
y_test_pred = classifier.predict(X_test_transformed)
```

```python
# evaluating
metrics.accuracy_score(y_test, y_test_pred)
```
